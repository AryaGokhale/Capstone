```python
def extract_pe_information(file_path):
    try:
        pe = pefile.PE(file_path)

        # Address of Entry Point
        address_of_entry_point = hex(pe.OPTIONAL_HEADER.AddressOfEntryPoint)

        # Image Base
        image_base = hex(pe.OPTIONAL_HEADER.ImageBase)

        # Linker Version
        major_linker_version = pe.OPTIONAL_HEADER.MajorLinkerVersion
        minor_linker_version = pe.OPTIONAL_HEADER.MinorLinkerVersion

        # Operating System Version
        major_os_version = pe.OPTIONAL_HEADER.MajorOperatingSystemVersion
        minor_os_version = pe.OPTIONAL_HEADER.MinorOperatingSystemVersion

        # Image Version
        major_image_version = pe.OPTIONAL_HEADER.MajorImageVersion
        minor_image_version = pe.OPTIONAL_HEADER.MinorImageVersion

        # Size of Code
        size_of_code = pe.OPTIONAL_HEADER.SizeOfCode

        # Size of Initialized Data
        size_of_initialized_data = pe.OPTIONAL_HEADER.SizeOfInitializedData

        # Size of Uninitialized Data
        size_of_uninitialized_data = pe.OPTIONAL_HEADER.SizeOfUninitializedData

        # Size of Image
        size_of_image = pe.OPTIONAL_HEADER.SizeOfImage

        # Size of Headers
        size_of_headers = pe.OPTIONAL_HEADER.SizeOfHeaders

        # Subsystem
        subsystem = pe.OPTIONAL_HEADER.Subsystem

        # DllCharacteristics
        dll_characteristics = pe.OPTIONAL_HEADER.DllCharacteristics

        # Size of Stack Reserve and Commit
        size_of_stack_reserve = pe.OPTIONAL_HEADER.SizeOfStackReserve
        size_of_stack_commit = pe.OPTIONAL_HEADER.SizeOfStackCommit

        # Size of Heap Reserve and Commit
        size_of_heap_reserve = pe.OPTIONAL_HEADER.SizeOfHeapReserve
        size_of_heap_commit = pe.OPTIONAL_HEADER.SizeOfHeapCommit

        # Return the extracted information
        return {
            "AddressOfEntryPoint": address_of_entry_point,
            "ImageBase": image_base,
            "MajorLinkerVersion": major_linker_version,
            "MinorLinkerVersion": minor_linker_version,
            "MajorOperatingSystemVersion": major_os_version,
            "MinorOperatingSystemVersion": minor_os_version,
            "MajorImageVersion": major_image_version,
            "MinorImageVersion": minor_image_version,
            "SizeOfCode": size_of_code,
            "SizeOfInitializedData": size_of_initialized_data,
            "SizeOfUninitializedData": size_of_uninitialized_data,
            "SizeOfImage": size_of_image,
            "SizeOfHeaders": size_of_headers,
            "Subsystem": subsystem,
            "DllCharacteristics": dll_characteristics,
            "SizeOfStackReserve": size_of_stack_reserve,
            "SizeOfStackCommit": size_of_stack_commit,
            "SizeOfHeapReserve": size_of_heap_reserve,
            "SizeOfHeapCommit": size_of_heap_commit
        }

    except Exception as e:
        print(f"Error: {e}")
        return None

# Example usage:
file_path = "/content/DikeDataset/files/malware/000077419ead44800537b34f5c2137e572c472698a13c39151f90eaabde7c94e.exe"
pe_information = extract_pe_information(file_path)
if pe_information:
    for key, value in pe_information.items():
        print(f"{key}: {value}")
```

```python
from capstone import *
from collections import Counter

def disassemble_code_section(pe):
    disassembled_code = []

    # Initialize Capstone disassembler
    md = Cs(CS_ARCH_X86, CS_MODE_32)

    # Iterate through each section
    for section in pe.sections:
        if section.Characteristics & 0x20:  # IMAGE_SCN_CNT_CODE
            code = section.get_data()
            addr = section.VirtualAddress

            # Disassemble each instruction in the code section
            for insn in md.disasm(code, addr):
                disassembled_code.append((insn.address, insn.mnemonic, insn.op_str))

    return disassembled_code

# def extract_opcode_info(file_path):
#     try:
#         pe = pefile.PE(file_path)
#         opcodes = disassemble_code_section(pe)
#         return opcodes
#     except Exception as e:
#         print(f"Error: {e}")
#         return None

# Example usage:
# file_path = "/content/DikeDataset/files/malware/000077419ead44800537b34f5c2137e572c472698a13c39151f90eaabde7c94e.exe"
# opcode_info = extract_opcode_info(file_path)
# if opcode_info:
#     for addr, mnemonic, op_str in opcode_info:
#         print(f"Address: {hex(addr)}, Mnemonic: {mnemonic}, Operands: {op_str}")

def extract_opcode_info(file_path):
    opcode_frequency = Counter()  # Counter to store opcode frequency
    try:
        pe = pefile.PE(file_path)
        opcodes = disassemble_code_section(pe)

        # Count opcode frequency
        for _, mnemonic, op_str in opcodes:
            opcode_frequency[(mnemonic, op_str)] += 1

        return opcode_frequency
    except Exception as e:
        print(f"Error: {e}")
        return None

def save_to_csv(file_path, opcode_frequency):
    with open(file_path, 'w', newline='') as csvfile:
        fieldnames = ['Mnemonic', 'Operands', 'Frequency']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for (mnemonic, op_str), freq in opcode_frequency.items():
            writer.writerow({'Mnemonic': mnemonic, 'Operands': op_str, 'Frequency': freq})

# Example usage:
file_path = "/content/DikeDataset/files/malware/000077419ead44800537b34f5c2137e572c472698a13c39151f90eaabde7c94e.exe"
opcode_frequency = extract_opcode_info(file_path)
if opcode_frequency:
    for (mnemonic, op_str), freq in opcode_frequency.items():
        print(f"Mnemonic: {mnemonic}, Operands: {op_str}, Frequency: {freq}")

csv_path = "/content/000077419ead44800537b34f5c2137e572c472698a13c39151f90eaabde7c94e.csv"
save_to_csv(csv_path, opcode_frequency)
```
